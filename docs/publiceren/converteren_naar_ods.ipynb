# Converteer meldingen data naar ODS meldingen data
Op basis van https://data.overheid.nl/dataset/mor-slim-melden-groningen



# importeren van data

```python
importfile = 'https://ckan.dataplatform.nl/dataset/885dbda5-efc1-4d64-b1d6-8444cd5a4cb8/resource/edc8db66-8dc3-4819-9935-e6c8f55388ab/download/meldingen_openbare_ruimte_groningen.csv'
exportfile = 'converted-mor-data-ods.csv'
```

```python
import pandas as pd
df =  pd.read_csv(importfile, engine="python", sep=',', quotechar='"', error_bad_lines=False)

print(len(df),'rows import as job') 
```

# Converteren van de data
*Bron:
zaaknummer status        lat       lon  begindatum einddatum                         datasetnaam categorie  aantal_stemmen  description media1 media2 media3  
*Target: File, identificatie, publicatie, hoofdcategorie,subcategorie, omschrijving,status,datum_melding, einddatum,latidude,longitude, media1,media2,media3,kanaal, "BOR_MELD"

```python
df.head()
```

```python


ndf = df.drop(columns=['datasetnaam', 'aantal_stemmen'])

ndf = ndf.rename(columns={
    "zaaknummer": "identificatie",
        "lat": "latidude",
        "lon": "longitude",
        "begindatum": "datum_melding",
        "categorie": "hoofdcategorie",
     "description": "omschrijving"
})

ndf.head()
```

```python
print(len(ndf),'rows exported as ods result') 
ndf.to_csv(exportfile, index=False )
```
